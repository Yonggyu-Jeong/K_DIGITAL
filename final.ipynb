```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import plot_confusion_matrix
import seaborn as sns

pd.set_option('display.max_rows', None)
df_test = pd.read_csv("C:\\Users\\wjddy\\Documents\\카카오톡 받은 파일\\FordA_TEST.csv")
df_train = pd.read_csv("C:\\Users\\wjddy\\Documents\\카카오톡 받은 파일\\FordA_TRAIN.csv")
```

```python
attributeList = []
for i in range(1, 501):
    attributeList.append("att"+str(i))
    

```

```python
df_test_Series = df_test[attributeList].sum() / df_test[attributeList].count()
```

```python
df_train_Series = df_train[attributeList].sum() / df_train[attributeList].count()
```

```python

fig = plt.figure(figsize=(16,8)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성
 
ax.plot(df_test_Series ,marker='o',label='df_test', color='red') ## 선그래프 생성
ax.plot(df_train_Series ,marker='o',label='df_train', color='blue') 
 
ax.legend() ## 범례
 
plt.show()
```
